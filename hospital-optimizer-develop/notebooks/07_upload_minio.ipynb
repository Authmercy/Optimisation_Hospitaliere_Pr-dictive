```python
# Installation du client MinIO 
import subprocess
subprocess.run(["pip", "install", "minio", "--quiet"])
print("✅ Client MinIO installé")
```

```python
# Upload des 26 fichiers
from minio import Minio
import os

# Connexion MinIO
client = Minio(
    "hospital-minio:9000",
    access_key="hospital",
    secret_key="hospital123",
    secure=False
)

# Vérifier que le bucket bronze existe
if not client.bucket_exists("bronze"):
    client.make_bucket("bronze")
    print("✅ Bucket bronze créé")

# Dossier source
SOURCE_DIR = "/home/jovyan/data/raw/mimic_synthetic"

# Upload de chaque fichier
files = sorted(os.listdir(SOURCE_DIR))
print(f"Fichiers à uploader : {len(files)}\n")

for i, filename in enumerate(files, 1):
    filepath = os.path.join(SOURCE_DIR, filename)
    size_mb  = os.path.getsize(filepath) / (1024**2)

    print(f"[{i:02d}/{len(files)}] Upload {filename} ({size_mb:.1f} MB)...")

    client.fput_object(
        "bronze",
        f"mimic_synthetic/{filename}",
        filepath
    )
    print(f"          ✅ OK")

print("\n✅ Tous les fichiers uploadés dans MinIO bucket bronze")
```
